```python
def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks

tracks = get_playlist_tracks("spotify", "7jISe26ax1OT0xfXqC32DE")
```

```python
def get_artists_from_playlist(playlist_id):
    
    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)
    
    artists = []
    
    for track in tracks_from_playlist:
        artists_info = track['track']['artists']
        
        for artist_info in artists_info:
            artists.append(artist_info['name'])
    
    return list(set(artists))
```

```python
def get_artists_ids_from_playlist(playlist_id):
    
    tracks_from_playlist = get_playlist_tracks("spotify", playlist_id)
    
    artists_ids = []
    
    for track in tracks_from_playlist:
        artists_info = track['track']['artists']
        
        for artist_info in artists_info:
            artists_ids.append(artist_info['id'])
            
    return list(set(artists_ids))
```

```python
artists_ids = get_artists_ids_from_playlist("7jISe26ax1OT0xfXqC32DE")
artists_ids
```

```python
def get_albums_from_artist(playlist_id):
    artists_ids_from_playlist = get_artists_ids_from_playlist(playlist_id)
    artist_album_names = []
    artist_album_uris = []
    for artist in artists_ids_from_playlist:
        results = sp.artist_albums(artist, album_type='album')
        for i in range(len(results['items'])):
            artist_album_names.append(results['items'][i]['name'])
            artist_album_uris.append(results['items'][i]['uri'])
    return list(set(artist_album_names)), list(set(artist_album_uris))
```

```python
def get_all_tracks_from_all_albums_from_all_artists_playlist(playlist):
    artist_album_names, artist_album_uris = get_albums_from_artist(playlist)
    all_tracks = []
    for album in artist_album_uris:
        tracks = sp.album_tracks(album)
        all_tracks.extend(tracks['items'])
    return all_tracks
```

```python
def get_all_track_ids(all_tracks):
    ids = []
    for track in all_tracks:
        ids.append(track['id'])
    return ids
```

```python
def getTrackFeatures(id):
    meta = sp.track(id)
    features = sp.audio_features(id)
    name = meta['name']
    album = meta['album']['name']
    artist = meta['album']['artists'][0]['name']
    release_date = meta['album']['release_date']
    length = meta['duration_ms']
    popularity = meta['popularity']
    acousticness = features[0]['acousticness']
    danceability = features[0]['danceability']
    energy = features[0]['energy']
    instrumentalness = features[0]['instrumentalness']
    liveness = features[0]['liveness']
    loudness = features[0]['loudness']
    speechiness = features[0]['speechiness']
    tempo = features[0]['tempo']
    time_signature = features[0]['time_signature']
    track = [name, album, artist, release_date, length, popularity, danceability, acousticness, danceability, energy, instrumentalness, liveness, loudness, speechiness, tempo, time_signature]
    return track
```

```python
def get_all_tracks_info(ids):
    all_tracks_info = []
    for i in range(len(ids)):
        time.sleep(.5)
        all_tracks_info.append(getTrackFeatures(ids[i]))
    return all_tracks_info
```

```python
all_tracks_info = get_all_tracks_info(ids)
```

```python
df = pd.DataFrame(all_tracks_info, columns = ['name', 'album', 'artist', 'release_date', 'length', 'popularity', 'danceability', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'time_signature'])
```
